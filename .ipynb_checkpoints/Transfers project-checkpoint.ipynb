### Champions League winners (2000/2001 to 2018/2019) 

```python
link = "http://www.rsssf.com/tablese/ec1.html"
resp = requests.get(link)
content = str(resp.content)
soup = BeautifulSoup(content, 'html.parser')
links = soup.findAll('a',href = True)
championsLeagueWinners = pd.DataFrame()
seasons = []
winners = []
for link in links:
    season = link.text
    slashIndex = season.find("/")
    if(slashIndex != -1):
        seasonStart = int(season[:slashIndex])
        if seasonStart >= 2000 and seasonStart < 2019:
            seasonEnd = "20" + season[slashIndex + 1:]
            seasons.append(str(seasonStart) + "-" + seasonEnd)
            winnerTeam = unidecode(link.next_sibling.replace("\\r\\n","").strip())
            winners.append(winnerTeam)
        
championsLeagueWinners = pd.DataFrame({"Season" : seasons, "Winner": winners})
championsLeagueWinners
```

### Format
Position - TeamName - GamesPlayed - Wins - Draws - Defeats - Goals Scored and Goals Missed - Points

```python
def ReadOneSeason(link,season,seasonEndDate,count):
    link = link.replace("#seasonEndDate#",str(seasonEndDate))
    resp = requests.get(link)
    content = str(resp.content)
    soup = BeautifulSoup(content, 'html.parser')
    table = soup.findAll("pre")[0].text

    lines = table.split("\\r\\n")

    lines = [x.replace("-"," ").replace("\\t"," ").strip() for x in lines if x.replace("-","").strip()]
    lines = [unidecode(x) for x in lines]
    startIndex = lines.index("Final Table:") + 1
    endIndex = 0
    for i, line in enumerate(lines):
        indexes = [line.find('Round'),line.find('NB'), line.find('[*] p'), line.find('[*] a'), line.find('Promot'),line.find('[*] R') ]
        indexes = [i for i in indexes if i >= 0]
        if len(indexes) > 0:
            endIndex = i
            break
    if endIndex == 0:
        endIndex = startIndex + count
    lines = lines[startIndex : endIndex]
    lines = [' '.join(x.split()) for x in lines]
    
    teams = []
    wins = []
    draws = []
    lost = []
    goalsScored = []
    goalsMissed = []
    points = []

    for line in lines:
        teamInfo = line.split()
        positionName = teamInfo[0].split(".")

        teamName = positionName[len(positionName) - 1]

        for i in range(1,len(line)):
            if teamInfo[i].isnumeric() and teamInfo[i][0] != '0' and teamInfo[i]!= "1860" and teamInfo[i] != "96" and teamInfo[i] != "1899":
                numericIndex = i
                break
            else:
                teamName = teamName + " " + teamInfo[i]
        teams.append(teamName) 
        wins.append(int(teamInfo[numericIndex + 1]))
        draws.append(int(teamInfo[numericIndex + 2]))
        lost.append(int(teamInfo[numericIndex + 3]))
        goalsScored.append(int(teamInfo[numericIndex + 4]))
        goalsMissed.append(int(teamInfo[numericIndex + 5]))
        points.append(int(teamInfo[numericIndex + 6]))
    df = pd.DataFrame({"Team": teams, "Win": wins, "Draw": draws, "Lost": lost,"Goals Scored" : goalsScored,
                       "Goals Missed": goalsMissed, "Points" : points})
    df = df.sort_values("Points", ascending = False)
    df["Position"] = np.arange(1,df.shape[0] + 1)
    df["Season"] = season
    return df
```

```python
def ReadAllSeasons(link,count):
    frames = []
    for i in range(2000,2019):
        if i + 1 < 2010:
            endSeason = str(i + 1)[2:4]
        else:
            endSeason = i + 1
        
        frames.append(ReadOneSeason(link,str(i) + "-" + str((i + 1)), endSeason,count))
    df = pd.concat(frames)    
    return df
```
